```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

# Sample data
data = {
    "date_added": ["September 24, 2021", "October 15, 2020", "November 05, 2019", "August 12, 2020"],
    "video_id": [1, 2, 3, 4]  # Sample video_id data
}

# Create DataFrame
df2 = pd.DataFrame(data)

# Convert 'date_added' column to datetime with the correct format
df2["date_added"] = df2["date_added"].apply(lambda x: datetime.strptime(x, "%B %d, %Y"))

# Extract the year from 'date_added' and create a new column 'year'
df2['year'] = df2['date_added'].dt.year

# Group by 'year' and count the number of 'video_id' per year
yearly_counts = df2.groupby('year')['video_id'].count()

# Create a bar chart
yearly_counts.plot(kind='bar', xlabel='Year', ylabel='Total Movies', title='Total Released Movies Per Year')

# Show the plot
plt.show()
```

```python
df['release_year'] = pd.to_datetime(df['release_year'])
df['release_year'] = df['release_year'].dt.date
video_count_by_date = df.groupby('release_year').size()
plt.figure(figsize=(12, 6))
sns.lineplot(data=video_count_by_date)
plt.title('Release of movies in years')
plt.xlabel('Relased year')
plt.ylabel('Nimber of movies')
plt.xticks(rotation=90)
plt.show()
```
