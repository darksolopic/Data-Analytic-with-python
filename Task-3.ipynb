```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv(r"C:\\Users\\ASUS\Downloads\\householdtask3.csv")
data.head()
```

```python
display(data.head(10))
```

```python
#scatter plot used to obereve relationship between varraibles and uses dots to represent
#scatter plot with year against own
plt.scatter(data['year'], data['own'])

#adding title to the plot
plt.title("Scatter Plot")

#setting the x and y y lable
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#line chart used to represent a relationship between two datsets on a different axises
#line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])

#adding title to the plot
plt.title("Line Plot")

#setting the x and y y lable
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#Bar chart: data visualization used to represent data in the form of rectangular bars.
#The height of each bar represents the value of a data point,
#and the width of each bar represents the category of the data.

#Bar Chart  with year against own
plt.bar(data['year'], data['own'])

#adding title to the plot
plt.title("Bar Chart")

#setting the x and y y lable
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#Histogram histogram is a graph showing frequency distributions.
#It is a graph showing the number of observations within each given interval.
#used to represent data provided in the form of some groups.

plt.hist(data['income'])

plt.title("Histogram")

plt.show()
```
